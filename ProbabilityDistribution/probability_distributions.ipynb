```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

mu, sigma = 0, 0.1 # Mean and standard deviation
s = np.random.normal(mu, sigma, size=10000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.title("Normal Distribution", fontsize=16)
plt.show()
```

```python
s = np.random.uniform(-1,0,1000)
print(f'Mean of Sample: {s.mean()}')
count, bins, ignored = plt.hist(s, 15, density=True)
plt.title("Uniform Distribution", fontsize=16)
plt.show()
```

```python
s = np.random.standard_cauchy(1000000)
s = s[(s>-25) & (s<25)]  # truncate distribution so it plots well
plt.hist(s, bins=100, density=True)
plt.title("Cauchy Distribution", fontsize=16)
plt.show()
```

```python
# t Distribution with df = 1
s = np.random.standard_t(df =1, size=100000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("t Distribution (df=1)", fontsize=16)
plt.show()
```

```python
# dfnum: between group degrees of freedom
# dfden: within groups degrees of freedom
# F Distribution with dfnum=1, dfden=1
s = np.random.f(dfnum=1, dfden=1, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("F Distribution (dfnum=1, dfden=1)", fontsize=16)
plt.show()
```

```python
# Chi Square Distribution with df = 1
s = np.random.chisquare(df=1, size=10000)
s = s[np.where(abs(s) < 20)]
h = plt.hist(s, bins=100, density=True)
plt.title("Chi Square Distribution (df=1)", fontsize=16)
plt.show()
```

```python
# Exponential distribution with lamda=1
s = np.random.exponential(scale=1.0, size=10000)
h = plt.hist(s, bins=1000, density=True)
plt.title("Exponential Distribution", fontsize=16)
plt.show()
```

```python
# Weibull Distribution with gamma = 0.5
s = np.random.weibull(a=0.5, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("Weibull Distribution (gamma=0.5)", fontsize=16)
plt.show()
```

```python
#  LogNormal Distribution with sigma = 0.5
s = np.random.lognormal(mean=0.0, sigma=0.5, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("LogNormal Distribution (sigma=1)", fontsize=16)
plt.show()
```

```python
from scipy.stats import fatiguelife
# fatiguelife Distribution with gamma = 0.5
x = np.linspace(fatiguelife.ppf(0.01, c=0.5), fatiguelife.ppf(0.99, c=0.5), 100)
plt.plot(x, fatiguelife.pdf(x, c=0.5), 'b-', lw=2, alpha=0.6, label='fatiguelife pdf')
plt.title("fatiguelife Distribution (gamma=0.5)", fontsize=16)
plt.show()
```

```python
# Gamma Distribution with gamma = 0.5
s = np.random.gamma(shape=0.5, scale=1.0, size=10000)
s = s[np.where(abs(s) < 10)]
h = plt.hist(s, bins=100, density=True)
plt.title("Gamma Distribution (gamma=0.5)", fontsize=16)
plt.show()
```

```python
loc, scale = 0., 1.
s = np.random.laplace(loc, scale, 10000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.title("Laplace Distribution", fontsize=16)
plt.show()
```

```python
from scipy.stats import powernorm
# Power Normal Distribution with p = 0.5
x = np.linspace(powernorm.ppf(0.01, c=0.5),powernorm.ppf(0.99, c=0.5), 100)
plt.plot(x, powernorm.pdf(x, c=0.5), 'b-', lw=2, alpha=0.6, label='powernorm pdf')
plt.title("Power Normal Distribution (p=0.5)", fontsize=16)
plt.show()
```

```python
from scipy.stats import powerlognorm
# Powerlog Normal Distribution with p = 0.5
x = np.linspace(powerlognorm.ppf(0.01, c=0.5, s=1), powerlognorm.ppf(0.99, c=0.5, s=1), 100)
x = x[np.where(np.abs(x) < 5)]
plt.plot(x, powerlognorm.pdf(x, c=0.5, s=1), 'b-', lw=2, alpha=0.6)
plt.title("Powerlog Normal Distribution (p=0.5)", fontsize=16)
plt.show()
```

```python
from scipy.stats import tukeylambda
# tukeylambda Distribution with lambda = 0.5
lam = -1
x = np.linspace(tukeylambda.ppf(0.01, lam), tukeylambda.ppf(0.99, lam), 100)
plt.plot(x, tukeylambda.pdf(x, lam), 'b-', lw=2, alpha=0.6)
plt.title("Tukey Lambda Distribution (lambda=-1)", fontsize=16)
plt.show()
```

```python
mu, beta = 0, 0.1 # location and scale
s = np.random.gumbel(mu, beta, 1000)
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.title("Gumbel Distribution", fontsize=16)
plt.show()
```

```python
# Beta Distribution with (alpha=0.5, beta=0.5)
s = np.random.beta(a=0.5, b=0.5, size=10000)
plt.hist(s, bins=1000, density=True)
plt.title("Beta Distribution (alpha=0.5, beta=0.5)", fontsize=16)
plt.show()
```

```python
# Binomial Distribution with (n=100, p=0.1)
n, p = 100, .1  # number of trials, probability of each trial
s = np.random.binomial(n, p, 10000)
h = plt.hist(s, bins=20, density=True)
plt.title("Binomial Distribution (n=100, p=0.1)", fontsize=16)
plt.show()
```

```python
# Poisson Distribution with (lambda = 5)
s = np.random.poisson(lam=5, size=100000)
h = plt.hist(s, bins=20, density=True)
plt.title("Poisson Distribution (lambda = 5)", fontsize=16)
plt.show()
```
