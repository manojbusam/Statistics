```python
mean = df.height.mean()
mean
```

```python
# 1. Removing outliers using standard deviation
# 7 outliers are being removed below
df_no_outlier = df[(df.height < mean + 3*std_deviation) & (df.height > mean - 3*std_deviation)]
df_no_outlier.shape
```

```python
df['zscore'] = (df.height - df.height.mean())/df.height.std()
df.head()
```
